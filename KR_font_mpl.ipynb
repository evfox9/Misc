# Matplotlib에 한글 폰트 설치하기

먼저 필요한 패키지를 불러옵니다

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

설치되어 있는 폰트를 불러온 다음 한글을 지원하는 폰트를 찾습니다.

```python
font_list = fm.findSystemFonts(fontpaths = None, fontext = 'ttf')

print(font_list)
```

(어느 폰트가 한글을 지원하는지 모르겠는 경우) 
폰트 이름에 'cjk'가 들어가는 폰트는 전부 한글을 지원합니다.

```python
print([i for i in font_list if 'CJK' in i])
```

원하는 폰트의 주소를 복사하여 아래 코드를 작성해주시면 됩니다.

```python
font_location = '/usr/share/fonts/opentype/noto/NotoSansCJK-Light.ttc'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
```
